```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../Images/sea_beach.jpg")
b, g, r = cv2.split(img)
```

```python
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.hist(b.ravel(), 256, [0, 256])
plt.show()
cv2.imshow("b", b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.hist(g.ravel(), 256, [0, 256])
plt.show()
cv2.imshow("g", g)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.hist(r.ravel(), 256, [0, 256])
plt.show()
cv2.imshow("r", r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
